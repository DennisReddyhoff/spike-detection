```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime
from scipy.signal import savgol_filter
from scipy.stats import zscore
import haversine as hs
```

```python
df = pd.read_csv('data-warehouse-2021.csv')
```

```python
df.keys()
```

```python
tmp = df[['timestamp','device','site_longitude', 'site_latitude', 'pm2_5', 's2_pm2_5']].dropna()
```

```python
tmp['pm2_5'] = pd.to_numeric(tmp['pm2_5'])
```

```python
tmp['s2_pm2_5'] = pd.to_numeric(tmp['s2_pm2_5'])
```

```python
tmp = tmp[abs(tmp['pm2_5']-tmp['s2_pm2_5']) < 0.2*tmp['pm2_5']]
```

```python
tmp['timestamp'] = pd.to_datetime(tmp['timestamp'])
```

```python
tmp
```

```python
tmp['new_time'] = tmp['timestamp'].apply(lambda x: datetime.timestamp(x))
```

```python
tmp['new_time'] = tmp['new_time'] - tmp['new_time'].min()
```

```python
tmp
```

```python
tmp[tmp['device'] == 'aq_05'].plot(x='new_time', y=['pm2_5', 's2_pm2_5'])
```

```python
tmp[tmp['device'] == 'aq_09'].plot(x='new_time', y=['pm2_5', 's2_pm2_5'])
```

```python
aq_09_smooth = savgol_filter(tmp[tmp['device'] == 'aq_09']['pm2_5'], window_length = 21, polyorder = 3)
```

```python
plt.plot(tmp[tmp['device'] == 'aq_09']['new_time'], aq_09_smooth)
```

```python
plt.plot(tmp[tmp['device'] == 'aq_09']['new_time'],zscore(aq_09_smooth)>2)
```

```python
locations = tmp[['device', 'site_longitude', 'site_latitude']].drop_duplicates().reset_index(drop=True)
```

```python
locations
```

```python
nearest = []
for i in locations['device'].unique():
    dist = 999
    device = ''
    for j in locations['device'].unique():
        if i != j:
            point1 = (
                locations[locations['device'] == i]['site_latitude'].item(),
                locations[locations['device'] == i]['site_longitude'].item()
                     )
            point2 = (
                locations[locations['device'] == j]['site_latitude'].item(),
                locations[locations['device'] == j]['site_longitude'].item()
                     )
            hav = hs.haversine(point1,point2)
            if hav < dist:
                dist = hav
                device = j
            
    nearest.append((device, dist))
```

```python
locations['closest'] = [i[0] for i in nearest]
locations['distance'] = [i[1] for i in nearest]
```

```python
colocated = locations[locations['distance'] < 0.5].reset_index(drop=True)
```

```python
tmp[tmp['device'] == 'aq_05'].plot(x='new_time', y=['pm2_5', 's2_pm2_5'])
tmp[tmp['device'] == 'aq_68'].plot(x='new_time', y=['pm2_5', 's2_pm2_5'])
```

```python
aq_05_smooth = savgol_filter(tmp[tmp['device'] == 'aq_05']['pm2_5'], window_length = 21, polyorder = 3)
aq_68_smooth = savgol_filter(tmp[tmp['device'] == 'aq_68']['pm2_5'], window_length = 21, polyorder = 3)
```

```python
plt.plot(tmp[tmp['device'] == 'aq_05']['new_time'],zscore(aq_05_smooth)>2.5)
plt.plot(tmp[tmp['device'] == 'aq_68']['new_time'],zscore(aq_68_smooth)>2.5)
```

```python
tmp[tmp['device'] == 'aq_05'][tmp.timestamp.between('01-01-2021', '02-01-2021')].plot(x='new_time', y=['pm2_5', 's2_pm2_5'])
tmp[tmp['device'] == 'aq_68'][tmp.timestamp.between('01-01-2021', '02-01-2021')].plot(x='new_time', y=['pm2_5', 's2_pm2_5'])
```

```python
plt.plot(
    tmp[tmp['device'] == 'aq_05'][tmp.timestamp.between('01-01-2021', '02-01-2021')]['new_time'],
    savgol_filter(
        tmp[tmp['device'] == 'aq_05'][tmp.timestamp.between('01-01-2021', '02-01-2021')]['pm2_5'], 
        window_length = 21, 
        polyorder = 3
    )
)

plt.plot(
    tmp[tmp['device'] == 'aq_68'][tmp.timestamp.between('01-01-2021', '02-01-2021')]['new_time'],
    savgol_filter(
        tmp[tmp['device'] == 'aq_68'][tmp.timestamp.between('01-01-2021', '02-01-2021')]['pm2_5'], 
        window_length = 21, 
        polyorder = 3
    )
)
```
